# Определение уровня сложности английского по субтитрам

Просмотр фильмов на оригинальном языке - это популярный и действенный метод прокачаться при изучении иностранных языков. Важно выбрать фильм, который подходит студенту по уровню сложности, чтоб студент понимал 50 - 70 % диалогов. Чтобы выполнить это условие, преподаватель должен посмотреть фильм и решить, какому уровню он соответствует. Однако это требует больших временных затрат.

Наша задача: разработать ML решение для автоматического определения уровня сложности англоязычных фильмов.

В нашем распоряжении имеется 270 предварительно размеченных файлов, также в результате поиска по тематическим страницам в интернете дополнено несколько сотен субтитров, для которых экспертно указан уровень языка.

### Подготовка необходимых функций

Используем регулярные выражения из библиотеки Python re для выполнения различных задач предварительной обработки.
Выполним следующие преобразования:
- приведём все символы к нижнему регистру,
- заменим одним пробелом один или несколько пробелов,
- удалим все стоп-слова,
- удалим все несловесные символы, такие как специальные символы, цифры и т. д.,
- все одиночные символы, имеющие пробелы с обеих сторон, заменим одним пробелом,
- один символ из начала документа заменим одним пробелом,
- заменим одним пробелом один или несколько пробелов,
- выполним лемматизацию.

```python
# Комментарии на английском языке, поэтому загрузим стоп-слова для английского языка
stopwords = set(nltk_stopwords.words('english'))
stop_words = list(stopwords)
```

```python
# Для выполнения лемматизации
# инициализируем модель spacy 'для английского, оставив только компонент, необходимый для лемматизации
nlp = spacy.load('en_core_web_sm', disable=['parser', 'ner'])
```

```python
# Определим функцию для предобработки текста
def text_preprocessor(text):

    # Приведём все символы к нижнему регистру
    out_text = text.lower()

    # Удалим стоп-слова
    out_text = out_text.split()
    clean_tokens = tuple(map(lambda x: '' if x in stop_words else x, out_text))
    out_text = ' '.join(clean_tokens)

    # Удалим слова, в которых:
    # - по три одинаковых символа подряд и более,
    # - два одинаковых символа подряд, всего символов три
    # - два одинаковых символа подряд повторяется дважды
    out_text = out_text.split()
    clean_tokens = tuple(
        map(
            lambda x: ''
            if (re.search(r'([a-z])\1{2,}', x))
            or (re.search(r'\s([a-z])\1{1}[a-z]{1}\s|\s[a-z]{1}([a-z])\2{1}\s', x))
            or (re.search(r'([a-z])\1{1}([a-z])\2{1}', x))
            else x,
            out_text
        )
    )
    out_text = ' '.join(clean_tokens)

    # Удалим все символы, которые не являются буквами латинского алфавита
    out_text = re.sub(r'[^a-zA-Z]', ' ', out_text)

    # Выполним лемматизацию
    # Parse the sentence using the loaded 'en' model object `nlp`
    doc = nlp(out_text)
    # Extract the lemma for each token and join
    out_text = ' '.join([token.lemma_ for token in doc])

    # Все одиночные символы, имеющие пробелы с обеих сторон, заменим пробелом
    out_text = re.sub(r'\s+[a-z]\s+', ' ', out_text)

    # Все последовательности пробелов заменим пробелом
    out_text = ' '.join(out_text.split())

    return out_text
```

```python
# Предобработаем субтитры созданной функцией
df['subtitles'] = df['subtitles'].apply(text_preprocessor)
```

```python
# Определим трансформер для векторизации и добавления столбцов со скоростью речи,
# частотой слов из словаря Oxford (векторизация Bag of words)
class bow_transformer(BaseEstimator, TransformerMixin):
    def __init__(self, max_df):
        self.max_df = max_df
        self._estimator = CountVectorizer(encoding='utf8',
                                          preprocessor=None,
                                          max_df=self.max_df)
    def fit(self, X, y=None):
        _ = self._estimator.fit(X['subtitles'])
        return self

    def transform(self, X):
        X_out = self._estimator.transform(X['subtitles'])
        X_out.resize((X_out.shape[0], X_out.shape[1]+4))
        X_out = X_out.tolil()
        X_out[:,-4:] = X[['speech_velocity', 'C1_all', 'C1_unique', 'levels_unique_sum']].values
        X_out.tocsr()
        return X_out
```

```python
# Определим трансформер для векторизации и добавления столбцов со скоростью речи,
# частотой слов из словаря Oxford (векторизация TF-IDF)
class tfidf_transformer(BaseEstimator, TransformerMixin):
    def __init__(self, max_df, norm):
        self.max_df = max_df
        self.norm = norm
        self._estimator = TfidfVectorizer(encoding='utf8',
                                          preprocessor=None,
                                          max_df=self.max_df,
                                          norm=self.norm)
    def fit(self, X, y=None):
        _ = self._estimator.fit(X['subtitles'])
        return self

    def transform(self, X):
        X_out = self._estimator.transform(X['subtitles'])
        X_out.resize((X_out.shape[0], X_out.shape[1]+4))
        X_out = X_out.tolil()
        X_out[:,-4:] = X[['speech_velocity', 'C1_all', 'C1_unique', 'levels_unique_sum']].values
        X_out.tocsr()
        return X_out
```

### Проверка классификации по частоте использования слов из соответствующих словарей

```python
# Загрузим словари для различных уровней языка
oxford = []
try:
    oxford.append(open('/content/drive/MyDrive/Colab Notebooks/English_level/The_Oxford_3000_by_CEFR_level.pdf', 'rb'))
    oxford.append(open('/content/drive/MyDrive/Colab Notebooks/English_level/The_Oxford_5000_by_CEFR_level.pdf', 'rb'))
    oxford.append(open('/content/drive/MyDrive/Colab Notebooks/English_level/American_Oxford_3000_by_CEFR_level.pdf', 'rb'))
    oxford.append(open('/content/drive/MyDrive/Colab Notebooks/English_level/American_Oxford_5000_by_CEFR_level.pdf', 'rb'))
except:
    oxford.append(open('English_level/The_Oxford_3000_by_CEFR_level.pdf', 'rb'))
    oxford.append(open('English_level/The_Oxford_5000_by_CEFR_level.pdf', 'rb'))
    oxford.append(open('English_level/American_Oxford_3000_by_CEFR_level.pdf', 'rb'))
    oxford.append(open('English_level/American_Oxford_5000_by_CEFR_level.pdf', 'rb'))
```

```python
# Прочитаем слова из файлов со списками слов, распределив слова по спискам
# с соответствующим уровнем
levels = ['A1', 'A2', 'B1', 'B2', 'C1']
words_by_level = {'A1':[], 'A2':[], 'B1':[], 'B2':[], 'C1':[]}
# Прочитаем файл
for file in oxford:
    words = []
    pdf_reader = PyPDF2.PdfReader(file)
    # На каждой страницее файла
    for pdf_page in pdf_reader.pages:
        # Будем извлекать текст
        text = pdf_page.extract_text()
        # Для каждой строки в тексте страницы
        for string in text.split('\n'):
            # Если это не колонтитул
            if not ('Oxford' in string):
                # Очистим строку от цифр (кроме обозначения уровня)
                if not string in levels:
                    string = re.sub(r'\d', ' ', string)
                    string = string.lower()
                words_in_string = string.split()
                word_number = 0
                # В список слов файла добавим все слова строки
                # Слова, обозначающие часть речи и проч не будем добавлять
                while words_in_string[word_number][-1] == ',':
                    words.append(words_in_string[word_number][:-1])
                    word_number += 1
                words.append(words_in_string[word_number])
    # Распределим слова по отдельным спискам
    if 'A1' in words:
        words_by_level['A1'] += words[words.index('A1')+1:words.index('A2')]
    if 'A2' in words:
        words_by_level['A2'] += words[words.index('A2')+1:words.index('B1')]
    if 'B1' in words:
        words_by_level['B1'] += words[words.index('B1')+1:words.index('B2')]
    if 'C1' in words:
        words_by_level['B2'] += words[words.index('B2')+1:words.index('C1')]
        words_by_level['C1'] += words[words.index('C1')+1:]
    elif 'B2' in words:
        words_by_level['B2'] += words[words.index('B2')+1:]

# Избавимся от дублей
words_by_level['A1'] = list(set(words_by_level['A1']))
words_by_level['A2'] = list(set(words_by_level['A2']))
words_by_level['B1'] = list(set(words_by_level['B1']))
words_by_level['B2'] = list(set(words_by_level['B2']))
words_by_level['C1'] = list(set(words_by_level['C1']))
```

Добавим колонки с частотой встречаемости слов из словаря соответствующего уровня (сколько суммарно слов определённого уровня, а также сколько уникальных слов определённого уровня).

```python
for level in levels:
    vect = CountVectorizer(encoding='utf8',
                           preprocessor=None,
                           vocabulary = words_by_level[level]
                           )
    tf_matrix = vect.fit_transform(df['subtitles'])
    df[level+'_all'] = tf_matrix.sum(axis=1)
    unique_words = []
    for i in range(tf_matrix.shape[0]):
        unique_words.append(tf_matrix[i].count_nonzero())
    df[level+'_unique'] = unique_words
```

```python
df['levels_all_sum'] = df[['A1_all', 'A2_all', 'B1_all', 'B2_all', 'C1_all']].sum(axis=1)
df['levels_unique_sum'] = df[['A1_unique', 'A2_unique', 'B1_unique', 'B2_unique', 'C1_unique']].sum(axis=1)
for level in ['A1_all', 'A1_unique', 'A2_all', 'A2_unique', 'B1_all',
              'B1_unique', 'B2_all', 'B2_unique', 'C1_all', 'C1_unique']:
    if 'all' in level:
        df[level+'_rel'] = df[level]/df['levels_all_sum']
    else:
        df[level+'_rel'] = df[level]/df['levels_unique_sum']
```

### Корректировка процесса обучения и получение новых результатов

```python
# Переопределим трансформер для векторизации и добавления столбцов, оставив только скорость речи
class bow_transformer_new(BaseEstimator, TransformerMixin):
    def __init__(self, max_df):
        self.max_df = max_df
        self._estimator = CountVectorizer(encoding='utf8',
                                          preprocessor=None,
                                          max_df=self.max_df)
    def fit(self, X, y=None):
        _ = self._estimator.fit(X['subtitles'])
        return self

    def transform(self, X):
        X_out = self._estimator.transform(X['subtitles'])
        X_out.resize((X_out.shape[0], X_out.shape[1]+1))
        X_out = X_out.tolil()
        X_out[:,-1] = X['speech_velocity'].values
        X_out.tocsr()
        return X_out
```

```python
# Переопределим трансформер для векторизации и добавления столбцов, оставив только скорость речи
class tfidf_transformer_new(BaseEstimator, TransformerMixin):
    def __init__(self, max_df, norm):
        self.max_df = max_df
        self.norm = norm
        self._estimator = TfidfVectorizer(encoding='utf8',
                                          preprocessor=None,
                                          max_df=self.max_df,
                                          norm=self.norm)
    def fit(self, X, y=None):
        _ = self._estimator.fit(X['subtitles'])
        return self

    def transform(self, X):
        X_out = self._estimator.transform(X['subtitles'])
        X_out.resize((X_out.shape[0], X_out.shape[1]+1))
        X_out = X_out.tolil()
        X_out[:,-1] = X['speech_velocity'].values
        X_out.tocsr()
        return X_out
```

#### Логистическая регрессия с применением Bag of words
